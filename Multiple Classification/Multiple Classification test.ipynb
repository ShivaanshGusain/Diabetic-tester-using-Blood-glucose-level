```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("flipper_species_classification.csv")
x = df[['Flipper length (x)']]
y = df[['Species (y)']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.5, random_state = 42)
```

Using Multinomial Algorithm

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_pred, y_test)
print("Accuracy: ", accuracy)
df = pd.DataFrame({'Flipper length (x)': x_test['Flipper length (x)'], 'Species (y)': y_test['Species (y)']})
print(df)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Flipper length (x)'], df['Species (y)'], 
            c=df['Species (y)'], cmap='bwr', edgecolor='k')
plt.xlabel('Flipper Species')
plt.ylabel('Species ( 0,1,2 )')
plt.title('Flipper length Vs Species')
plt.grid(True)
plt.show()
```

Using OVR Algorithm 

```python
model = LogisticRegression(multi_class = 'ovr', max_iter = 300)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_pred, y_test)
print("Accuracy: ", accuracy)
```
