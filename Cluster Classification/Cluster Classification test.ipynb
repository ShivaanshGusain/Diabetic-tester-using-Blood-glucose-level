```python
# Re-run the previous code after state reset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample flower data: number of leaves (x1) and number of petals (x2)
data = {
    'Leaves (x1)': [0, 0, 1, 1, 1, 1, 2, 2, 2],
    'Petals (x2)': [5, 6, 3, 3, 6, 8, 3, 7, 8]
}

df = pd.DataFrame(data)

# Create feature array
X = df[['Leaves (x1)', 'Petals (x2)']]

# Apply K-Means clustering with k=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Plotting the clusters
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']
for i in range(3):
    clustered_data = df[df['Cluster'] == i]
    plt.scatter(clustered_data['Leaves (x1)'], clustered_data['Petals (x2)'], 
                label=f'Cluster {i}', color=colors[i])

# Plot centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='X', c='black', label='Centroids')

plt.title("K-Means Clustering of Flowers")
plt.xlabel("Leaves (x1)")
plt.ylabel("Petals (x2)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv('clustered_flower_data.csv')

x = df[['Leaves (x1)','Petals (x2)']]

kmeans = KMeans(n_clusters=3, random_state=42)
df['Clusters'] = kmeans.fit_predict(x)

plt.figure(figsize=(max(df['Leaves (x1)']),max(df['Petals (x2)'])))
colors = ['red','blue','green']
for i in range(3):
    clustered_data = df[df['Clusters'] == i]
    plt.scatter(clustered_data['Leaves (x1)'], clustered_data['Petals (x2)'], label=f'Cluster {i}', color=colors[i])
    
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, marker='X', c='black', label='Centroids')

plt.title("K-Means Clustering of Flowers")
plt.xlabel("Leaves (x1)")
plt.ylabel("Petals (x2)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```
