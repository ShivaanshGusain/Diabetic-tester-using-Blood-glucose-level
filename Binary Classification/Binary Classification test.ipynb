Importing Model and modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Reading the data

```python
data = pd.read_csv('blood_glucose_dataset.csv')
train_data = data[:150]
test_data = data[150:]
```

```python
x = train_data[['Blood glucose (x)']]
y = train_data[['Diabetic? (y)']]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

Below there is no reason to do this, since our data is not in the text form and its in numeric form.

```python
# vectorizer = TfidfVectorizer(max_features = 5000)
# x_train_vec = vectorizer.fit_transform(x_train)
# x_test_vec = vectorizer.transform(x_test)
```

Instead we can fit the data directly in the model

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

Predicting the values and finding out the accuracy scores

```python
y_pred = model.predict(x_test)
Score = accuracy_score(y_test, y_pred)
print('Score: ', Score)
```

```python
x_value = test_data[['Blood glucose (x)']]
y_value = model.predict(x_value)

df = pd.DataFrame({'Blood glucose (x)': test_data['Blood glucose (x)'], 'Diabetic? (y)': y_value})
pd.set_option('display.max_rows', None)
print(df)
```

Testing the Model with input data - 

```python
x = int(input("Enter the Blood Glucose level: "))
x_df= pd.DataFrame([[x]], columns=['Blood glucose (x)'])
y = model.predict(x_df)
if y == 1:
    print("The person is Diabetic")
else:
    print("The person is not Diabetic")
    
```
